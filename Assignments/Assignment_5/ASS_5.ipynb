```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

A. Use Scatter plot, bar plot, Box plot and Histogram

```python
data = pd.read_csv("/content/car_crashes - car_crashes.csv")
data.head()
```

```python
plt.scatter(data.speeding,data.alcohol)
sns.set_style("whitegrid")
plt.show()
```

```python
plt.plot(np.arange(len(data.speeding)),data.speeding)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line plot")
plt.show()
```

```python
sns.distplot(data.ins_premium,kde = False)
plt.title('dist plot')
```

```python
plt.hist(data.total,bins=50)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title('histogram')
plt.show()
```

```python
plt.bar(data.alcohol,np.arange(len(data.alcohol)),width=0.5,label="alcohol")
plt.legend()
plt.title("bar graph")
plt.show()
```

```python
sns.kdeplot(data.not_distracted)
plt.title("KDE plot")
```

```python
sns.boxplot(x="total",y= 'alcohol' ,data=data, showmeans=True)
plt.show()
```

```python
plt.stem(data.total,data.not_distracted,linefmt=":",markerfmt="r",bottom=4)
```
