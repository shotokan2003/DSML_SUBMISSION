```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#import the data
sales = pd.read_excel("/content/supermarket_sales.xlsx")
sales.describe()
```

```python
sales.count()
```

```python
sales.info()
```

```python
sales.shape
```

```python
#Create a basic countplot to get the number of transaction for each product line
plt.figure(figsize=(15,5)) #used to make cahnges in the figure
sns.countplot(x="Product line",data = sales) #used to plot the graph
```

```python
# make it horizontal bar plot
sns.countplot(y="Product line",data = sales)
```

```python
# to get the count of two categories - Product and Gender
# hue
plt.figure(figsize=(15,5))
sns.countplot(x="Product line",data = sales,hue = "Gender")
```

```python
#using different color palette
plt.figure(figsize=(15,5))
sns.countplot(x="Product line",data = sales,hue = "Gender",palette="tab20") #use of palette is advised
```

```python
#change style using facecolor,linewidth and edge color
plt.figure(figsize=(15,5))
sns.countplot(x="Product line",data = sales,hue = "Gender",palette="tab20",facecolor=(1,1,1),edgecolor=sns.color_palette("bright",12))
```

```python
#plot a heatmap using matplotlib
#imshow() = this function is used for creating heatmap
data = np.random.random((10,10))
plt.imshow(data,cmap='hot',interpolation = 'nearest')
plt.title("HeatMap")
plt.colorbar()
plt.show()
```

```python
#implement boxplot
#create a basic boxplot on Total numeric variable
sns.boxplot(x="Total",data = sales,width = 2)
```

```python
# Create a boxplot on numeric variable by a categorical variable
sns.boxplot(y="Payment",x="Total",hue="Gender",data = sales,showmeans=True)
#showmeans gives the mean of the boxplot in the triangle representation in the graph as shown
```
